# `Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.`

## Matplotlib is a widely used Python library for creating static, animated, and interactive visualizations in a variety of formats. It provides a high-level interface for drawing attractive and informative statistical graphics, charts, plots, and other visual representations of data.






Matplotlib is used for many purposes:-

- Data Visualization: Matplotlib is used to create various types of plots and charts to visualize data and make it more understandable and interpretable


- Exploratory Data Analysis (EDA): Matplotlib is crucial for understanding the distribution, relationships, and patterns in your data during the exploratory phase of data analysis.


- Customization: Matplotlib provides extensive customization options, allowing you to tailor your plots to match your specific requirements, branding, or personal style.
 ## `And Many more`



`Scatter Plot: A two-dimensional plot that displays individual data points as dots. Used to visualize the relationship between two continuous variables.`

`Bar Plot: A chart that uses rectangular bars to represent data. Commonly used to compare categorical data or show the frequency distribution of a dataset.`


`Histogram: Displays the distribution of a continuous variable by dividing it into intervals (bins) and counting the frequency of observations within each bin.`


`Line Plot: A basic line chart that displays data points connected by straight lines. Useful for visualizing trends over time or continuous variables`


`Pie Chart: A circular chart divided into sectors, each representing a portion of a whole. Useful for showing proportions or percentages of categorical data.`

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

__Scatter Plot: A two-dimensional plot that displays individual data points as dots. Used to visualize the relationship between two continuous variables.__

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y,color='r')
plt.title("Scatter Plot",color='r')
plt.xlabel("X axis",color='b')
plt.ylabel("y axis",color='g')
plt.show
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:


For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)  # Top-left subplot
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1)
plt.title("Line Plot 1")

plt.subplot(2, 2, 2)  # Top-right subplot
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2)
plt.title("Line Plot 2")

plt.subplot(2, 2, 3)  # Bottom-left subplot
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3)
plt.title("Line Plot 3")

plt.subplot(2, 2, 4)  # Bottom-right subplot
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4)
plt.title("Line Plot 4")

# Adjust layout for better spacing
plt.tight_layout()


# Show the plots
plt.show()
```

# `Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.`

# Bar Plot: A chart that uses rectangular bars to represent data. Commonly used to compare categorical data or show the frequency distribution of a dataset

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit,color='b',alpha=0.3)
plt.show()
```

# `Q5: What is a box plot? Why is it used? Using the following data plot a box plot.`

# `A boxplot is a graph that gives a visual indication of how a data set's mean, median, mode, minimum, maximum and outlier values are spread out and compare to each other.`

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot([box1,box2],labels=['Box 1', 'Box 2'])

plt.title("Box Plot Example")

plt.xlabel("Boxes")
plt.ylabel("Values")
plt.show()
```
